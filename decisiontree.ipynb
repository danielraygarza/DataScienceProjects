## Creating Our Tree Models

```python
from sklearn.model_selection import train_test_split

X = S_df.drop(labels=['decade','level_1','track','artist','uri','target'], axis=1, inplace=False)
y = S_df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier

# instantiate the DecisionTreeClassifier model with criterion gini index
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=7, random_state=0)

# fit the model
clf_gini.fit(X_train, y_train)
```

```python
# predict the test set results with criterion gini index
y_pred_gini = clf_gini.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion gini index: {0:0.4f}'.format(accuracy_score(y_test, y_pred_gini)))
```

```python
y_pred_train_gini = clf_gini.predict(X_train)
print('Training set score: {:.4f}'.format(clf_gini.score(X_train, y_train)))
print('Test set score: {:.4f}'.format(clf_gini.score(X_test, y_test)))
```

## Hyper Parameter Optimization

```python
from sklearn import decomposition, datasets
from sklearn import tree
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

std_slc = StandardScaler()
pca = decomposition.PCA()
dec_tree = tree.DecisionTreeClassifier()
pipe = Pipeline(steps=[('std_slc', std_slc),
                           ('pca', pca),
                           ('dec_tree', dec_tree)])
n_components = list(range(1,X.shape[1]+1,1))
criterion = ['gini', 'entropy']
max_depth = [2,4,6,8,10,12]


parameters = dict(pca__n_components=n_components,
                      dec_tree__criterion=criterion,
                      dec_tree__max_depth=max_depth)
```

```python
clf_GS = GridSearchCV(pipe, parameters)
clf_GS.fit(X, y)
```

```python
print('Best Criterion:', clf_GS.best_estimator_.get_params()['dec_tree__criterion'])
print('Best max_depth:', clf_GS.best_estimator_.get_params()['dec_tree__max_depth'])
print('Best Number Of Components:', clf_GS.best_estimator_.get_params()['pca__n_components'])
print(); print(clf_GS.best_estimator_.get_params()['dec_tree'])
```

```python
pipe2 = Pipeline(steps=[('pca', pca),
                           ('dec_tree', dec_tree)])
max_depth2 = np.arange(1,51)
parameters2 = dict(pca__n_components=n_components,
                      dec_tree__criterion=criterion,
                      dec_tree__max_depth=max_depth2)
clf_GS2 = GridSearchCV(pipe2, parameters2)
clf_GS2.fit(X, y)
```

```python
print('Best Criterion:', clf_GS2.best_estimator_.get_params()['dec_tree__criterion'])
print('Best max_depth:', clf_GS2.best_estimator_.get_params()['dec_tree__max_depth'])
print('Best Number Of Components:', clf_GS2.best_estimator_.get_params()['pca__n_components'])
print(); print(clf_GS2.best_estimator_.get_params()['dec_tree'])
```

```python
pipe3 = Pipeline(steps=[('dec_tree', dec_tree)])
parameters3 = dict(dec_tree__criterion=criterion,
                      dec_tree__max_depth=max_depth2)
clf_GS3 = GridSearchCV(pipe3, parameters3)
clf_GS3.fit(X, y)
```

```python
print('Best Criterion:', clf_GS3.best_estimator_.get_params()['dec_tree__criterion'])
print('Best max_depth:', clf_GS3.best_estimator_.get_params()['dec_tree__max_depth'])
print(); print(clf_GS3.best_estimator_.get_params()['dec_tree'])
```

```python
y_pred_ent = clf_ent.predict(X_test2)
print('Model accuracy score with criterion gini index: {0:0.4f}'.format(accuracy_score(y_test2, y_pred_ent)))
```

```python
y_pred_train_ent = clf_ent.predict(X_train2)
print('Training set score: {:.4f}'.format(clf_ent.score(X_train2, y_train2)))
print('Test set score: {:.4f}'.format(clf_ent.score(X_test2, y_test2)))
```
